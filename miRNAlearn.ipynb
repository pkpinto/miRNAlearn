```python
%matplotlib inline

import matplotlib as mpl
mpl.rcParams['figure.figsize'] = '16, 4.5'
import matplotlib.pyplot as plt

mpl.rcParams['axes.labelsize'] = 18
mpl.rcParams['axes.facecolor'] = '#ffffff'
mpl.rcParams['axes.edgecolor'] = '#bcbcbc'
mpl.rcParams['grid.color'] = '0.95'
mpl.rcParams['legend.fontsize'] = 14
```

We only use two variables, MFE and P_MFE, to characterise the miRNAs. This gives us a very simple start point for the implementation of the various learning algorithms.

Check miRNAlearn.py for the implementation of the functions computing the values of the variables.

```python
import numpy as np
np.random.seed(0)
```

```python
from miRNAlearnIO import fasta_iter
mirnas = [s for t,s in fasta_iter('examples/csec_mirbase.fa')]
print(len(mirnas))
```

```python
import miRNAlearn
mfe_fold, mfe, gfe_fold, gfe = miRNAlearn.RNAfold(mirnas)
P = miRNAlearn.ground_state_prob(mfe, gfe)
```

```python
shuffled = list()
for s in mirnas:
    s = list(s)
    np.random.shuffle(s)
    shuffled.append(''.join(s))
    np.random.shuffle(s)
    shuffled.append(''.join(s))
```

```python
import miRNAlearn
mfe_fold_shuffled, mfe_shuffled, gfe_fold_shuffled, gfe_shuffled = miRNAlearn.RNAfold(shuffled)
P_shuffled = miRNAlearn.ground_state_prob(mfe_shuffled, gfe_shuffled)
```

```python
import numpy as np

data = np.concatenate((np.column_stack((mfe, P)),
                       np.column_stack((mfe_shuffled, P_shuffled))))
target = np.concatenate(([1] * len(mfe), [0] * len(mfe_shuffled)))
```

```python
indices = np.random.permutation(len(data))
cut = 160
train = {'X': data[indices[:-cut]], 'y': target[indices[:-cut]]}
test = {'X': data[indices[-cut:]], 'y': target[indices[-cut:]]}
```

```python
plt.scatter(train['X'][:,0], train['y'], c='Black')
```

```python
plt.scatter(train['X'][:,1], train['y'], c='Black')
```

```python
from mpl_toolkits.mplot3d import Axes3D
_a = plt.subplot(1, 1, 1, projection='3d')
_a.scatter(train['X'][:,0], train['X'][:,1], train['y'], c='Black')
```

### KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(train['X'], train['y'])
```

```python
plt.scatter(train['X'][:,0], train['y'], c='Black')
plt.scatter(train['X'][:,0], np.array(knn.predict_proba(train['X']))[:,1], c='White')
```

```python
from mpl_toolkits.mplot3d import Axes3D
_a = plt.subplot(1, 1, 1, projection='3d')
_a.scatter(train['X'][:,0], train['X'][:,1], train['y'], c='Black')
_a.scatter(train['X'][:,0], train['X'][:,1], np.array(knn.predict_proba(train['X']))[:,1], c='White')
```
